Python协程详解
协程是Python异步编程的核心概念。

协程是可以暂停和恢复执行的函数，它能在多个入口点之间切换执行。

```python
import asyncio

# 普通函数 vs 协程
def normal_function():
    return "普通函数"

async def coroutine_function():
    return "协程"

# 调用方式对比
print(normal_function())  # 直接调用

# 协程需要特殊方式调用
coro = coroutine_function()
print(f"协程对象: {coro}")  # <coroutine object coroutine_function at 0x...>
```

协程的await就类似于生成器的yield，但功能更强大。

协程的底层表示

每个协程在底层都有几个关键组件：

代码对象（Code Object）：包含要执行的字节码

帧对象（Frame Object）：保存执行状态（局部变量、执行位置等）

栈信息：函数调用栈的状态

当协程暂停时，Python会保存当前的帧对象。恢复时，重新加载这个帧对象，从上次暂停的位置继续执行。


*****恢复执行*****：

事件循环将协程标记为"就绪"

调度器选择这个协程继续执行

恢复之前保存的执行状态
